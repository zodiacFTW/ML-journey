```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X=np.random.rand(50)*2
Y=4+3*X+np.random.randn(50)

plt.scatter(X,Y,color='steelblue',s=40)
plt.xlabel("size(scaled)")
plt.ylabel("price(scaled)")
plt.title("our training data")
plt.grid(alpha=0.3)
plt.show()

print(X.shape)
print(Y.shape)
```

```python
X_b=np.c_[np.ones((50,1)),X]

theta=np.linalg.inv(X_b.T@X_b)@X_b.T@Y
print("θ₀ (intercept):",theta[0].round(3))
print("θ₁ (slope):",theta[1].round(3))

X_plot=np.array([[1,0],[1,2]])
y_plot=X_plot@theta

plt.scatter(X,Y,color="steelblue")
plt.plot([0,2],y_plot,color='r',lw=2)
plt.title("normal equation")
plt.xlabel("size")
plt.ylabel("price")
plt.grid(alpha=0.3)
plt.show()
```

```python
learning_rate=0.1
n_iterations=1000
m=len(X)

theta_gd=np.zeros(2)

cst_history=[]
for i in range(n_iterations):

    y_pred=X_b@theta_gd
    error=y_pred-Y
    cst=1/(2*m)*np.sum(error)**2
    cst_history.append(cst)


    gradient=1/m*X_b.T@error
    theta_gd-=learning_rate*gradient

print("θ₀ (intercept):", theta_gd[0].round(3))  
print("θ₁ (slope):    ", theta_gd[1].round(3))  

plt.plot(cst_history,color='purple')
plt.xlabel("iteration")
plt.yable("error squared mean")
plt.title("gradient descent converging")
plt.grid(alpha=0.3)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X.reshape(m,1),Y)
print("θ₀ (intercept):", model.intercept_.round(3))
print("θ₁ (slope):", model.coef_[0].round(3))
```
