```python
import numpy as np 
import matplotlib.pyplot as plt 

size=np.array([600,800,1000,1200,1500,1800,2000,2500])
price=np.array([25,35,45,55,68,80,90,112])

plt.scatter(size,price,color='steelblue',s=60)
plt.xlabel("size(sq ft)")
plt.ylabel("price(in lakhs)")
plt.title("housing data")
plt.grid(alpha=0.3)
plt.show()
```

```python
from numpy.polynomial import polynomial as P 
x=np.linspace(0,10,15)
y=2*x+1 +np.random.randn(15)*2
x_plot=np.linspace(0,10,200)
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

#underfit
c0=np.polyfit(x,y,0)
axes[0].scatter(x,y)
axes[0].plot(x_plot,np.polyval(c0,x_plot),'r')
axes[0].set_title("under fit (degree 0) \n high bias")

#goodfit
c1=np.polyfit(x,y,1)
axes[1].scatter(x,y)
axes[1].plot(x_plot,np.polyval(c1,x_plot),'g')
axes[1].set_title("good fit (Degree 1)\nlow bias low variance")

#overfit
c12=np.polyfit(x,y,12)
axes[2].scatter(x,y)
axes[2].plot(x_plot,np.polyval(c12,x_plot),'orange')
axes[2].set_title("over fit (Degree 12)\nlow bias high variance")

plt.tight_layout()
plt.show()
```
